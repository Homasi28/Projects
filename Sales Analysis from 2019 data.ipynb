### Here I am importing the necessary libraries

```python
import pandas as pd
import os
```

## Task 1 : merging 12 months of sales data into a single file

```python
# April 2019 sales data
files = [file for file in os.listdir('.\Sales_Data')]
all_months_data= pd.DataFrame()

for file in files:
    df= pd.read_csv("./Sales_Data/"+file)
    all_months_data= pd.concat([all_months_data, df])
    
all_months_data.to_csv("all_data.csv",index=False)
```

```python
# Read in updated dataFrame
all_data= pd.read_csv('all_data.csv')
all_data.head()
```

### Clean up the data

### Clean rows where value is NaN

```python
### Observing which column do have value NaN
nan_df= all_data[all_data.isna().any(axis=1)]
nan_df.head()

#### dropping all values equal to NaN

all_data= all_data.dropna(how="all")
all_data.head()
```

### Find 'Or' and delete it

```python
all_data= all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```

### Convert columns to the correct type

```python
all_data['Quantity Ordered']= pd.to_numeric(all_data['Quantity Ordered']) ## Make int
all_data['Price Each']= pd.to_numeric(all_data['Price Each']) ## Make float

all_data.head()
```

#### Augment data with additional colums

#### Task 2: add month column

```python
all_data['Month']= all_data['Order Date'].str[0:2]
all_data['Month']= all_data['Month'].astype('int32')
all_data.head()
```

### Task 3: Add a sales column

```python
all_data['Sales']= all_data['Quantity Ordered']*all_data['Price Each']
all_data.head()
```

### Task 4: Add a city column

```python
## using .apply() function

def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x)+ ' '+ get_state(x))

all_data.head()
```

##### What was the best month for sales? How much was earned that month?

```python
results = all_data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt

months= range(1,13)  ## months from January to December

plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```

### What city had the highest number of sales?

```python
results = all_data.groupby('City').sum()
results
```

```python
import matplotlib.pyplot as plt

cities= [city for city, df in all_data.groupby('City')] ## return cities in the same order when utilizin the group by function

plt.bar(cities, results['Sales'])
plt.xticks(cities,rotation='vertical',size= 8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('City name')
plt.show()

## San Fransicso had the highest number of sales
```

#### What time should we display to maximize likelihood of customer's buying product?

```python
all_data['Order Date']= pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hour']= all_data['Order Date'].dt.hour
all_data['Minute']= all_data['Order Date'].dt.minute

all_data.head()
```

```python
hours= [hour for hour, df in all_data.groupby('Hour')]

plt.plot(hours,all_data.groupby('Hour').count())
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.grid()
plt.show()

# Around 11 am (11:00) or 7pm (19:00)
```

### What products are most often sold together?

```python
df= all_data[all_data['Order ID'].duplicated(keep=False)]

df['Grouped']= df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x)) ## the tranform function joins order ID and Product together seprated by a comma

df= df[['Order ID', 'Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter

count= Counter()
for row in df['Grouped']:
    row_list= row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
for key, value in count.most_common(10):
    print(key,value)
```

### What product sold the most? Why do you think it sold the most?

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']

products= [product for product, df in product_group]

plt.bar(products, quantity_ordered)
plt.xticks(products, rotation='vertical', size=8)
plt.xlabel('Products Name')
plt.ylabel('Quantity ordered')

plt.show()
```

```python
prices= all_data.groupby('Product').mean()['Price Each']

fig, ax1= plt.subplots()

ax2= ax1.twinx()
ax1.bar(products, quantity_ordered, color='k')
ax2.plot(products,prices, 'b-')

ax1.set_xlabel('Products Name')
ax1.set_ylabel('Quantity Ordered', color='r')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products,rotation='vertical',size=8)

plt.show()
```
